1. What is NumPy, and why is it widely used in Python?
    - NumPy (Numerical Python) is a Python library used for numerical computing. It provides support for multidimensional arrays, mathematical functions, linear algebra, random number generation, and more.
     It's widely used because:

     - It's faster than standard Python lists.

     - Enables vectorized operations.

     - Acts as the foundation for libraries like Pandas, SciPy, Scikit-learn, etc.

2. How does broadcasting work in NumPy?

   -  Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes. It automatically expands the smaller array to match the shape of the larger one during operations without copying data.

3. What is a Pandas DataFrame?

6. What are the differences between NumPy arrays and Python lists?

  - NumPy arrays and Python lists are both used to store collections of data, but they differ significantly in terms of performance, functionality, and usage. A NumPy array is a powerful n-dimensional array object provided by the NumPy library, optimized for numerical and scientific computations. Unlike Python lists, which can hold elements of different data types, NumPy arrays require all elements to be of the same data type, enabling more efficient memory usage and faster processing. NumPy arrays support vectorized operations, allowing users to perform element-wise computations without using explicit loops, which significantly boosts performance. Additionally, NumPy provides a wide range of mathematical functions and operations that are not available for standard Python lists. While Python lists are more flexible and easier for general-purpose programming, NumPy arrays are preferred for numerical tasks, especially in data science, machine learning, and scientific computing due to their speed, efficiency, and functionality.

7. What is a heatmap, and when should it be used?

 - A heatmap is a color-coded 2D representation of data values. It's commonly used for:

    - Correlation matrices

    - Confusion matrices

    - Highlighting patterns or intensity in tabular data

8. What does the term “vectorized operation” mean in NumPy?

  - Vectorized operations allow you to apply operations to entire arrays without using explicit loops, making code faster, cleaner, and more Pythonic.

9. How does Matplotlib differ from Plotly?

   - Matplotlib and Plotly are both powerful Python libraries used for data visualization, but they differ in design, capabilities, and typical use cases.

   - Matplotlib is a static plotting library that is widely used for creating basic to advanced 2D plots like line graphs, bar charts, scatter plots, and histograms. It provides a high degree of control over every element of a plot, making it ideal for publication-quality visualizations. However, the plots generated with Matplotlib are static by default, meaning they don’t support interactivity unless additional tools (like mpld3 or plotly wrappers) are used.

   - Plotly, on the other hand, is a modern graphing library known for its rich interactive and web-ready visualizations. It allows users to create zoomable, hoverable, and clickable plots easily, which is particularly useful for dashboards and presentations. Plotly supports not just 2D charts but also 3D plots, maps, and animations. While Matplotlib focuses more on customization and static visual quality, Plotly emphasizes user experience and interactivity.

    - In summary, Matplotlib is better for detailed static plots with fine control, while Plotly excels in creating interactive, visually engaging plots with minimal effort.

1. How do you create a 2D NumPy array and calculate the sum of each row?

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(arr, axis=1)
print("Row-wise sum:", row_sums)
```

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Math': [90, 85, 88, 92],
    'Science': [88, 90, 85, 91],
    'English': [70, 80, 75, 78]
})

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

5. Generate a bar plot using Plotly.

8. Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [15, 18, 22, 28]

plt.plot(x, y1, label="Line 1", marker='o')
plt.plot(x, y2, label="Line 2", marker='s')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```

13. Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 15, 13, 17],
    'z': [5, 6, 7, 8]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title="3D Scatter Plot")
fig.show()
```
